```python
import time
import pandas as pd
import matplotlib.pyplot as plt
#mi complejidad es o(n) porque solo recorreré máximo una vez la lista
start=time.time()
def imprimirLista(lista):
	for element in lista:
		print(element)

listA = [1]
tiempos=[]
for num in range(1,31):
    start=time.time()
    imprimirLista(listA)
    end=time.time()
    final=end-start
    tiempos.append(final)
    listA.append(num)
print(tiempos)
indices = range(len(tiempos))
plt.bar(indices, tiempos)
```

```python
#mi complejidad es o(n) porque máximo solo checaré una vez todo mi input
import random
def balanceada(texto):
	stack=[]
	opening=["(","[","{"]
	closing=[")","]","}"]

	if(texto[0] in closing):
		return False
	for element in texto:
		if element in opening:
			stack.append(element)
		if(len(stack)!=0):
			if element==")":
				if(stack[-1]=="("):
					stack.pop()
			if element=="]":
				if(stack[-1]=="["):
					stack.pop()
			if element=="}":
				if(stack[-1]=="{"):
					stack.pop()
	if(len(stack)!=0):
		return False
	else:
		return True
palabra=["(","[","{", ")", "]", "}"]
lista=[]   
tiempos=[]
for num in range(1,30):
    lista.append(random.choice(palabra))
    random.shuffle(lista)
    start=time.time()
    print(lista)
    print(balanceada(lista))
    end=time.time()
    final=end-start
    tiempos.append(final)
print(tiempos)
indices = range(len(tiempos))
plt.bar(indices, tiempos)
```

```python
import string
from random import shuffle
#la complejid es o(n) porque máximo solo recorreré el tamaño de mi lista una vez
def anagrama(text1,text2):
	lst1=text1
	lst2=text2
	count=0
	for letter in text1:
		if(letter in text2):
			count+=1
	if(count==len(lst2)):
		return True
	else:
		return False

print(anagrama("diego","diegos"))
```

```python
# Complejidad: n
# Esto porque  en el peor de los casos, este siendo donde toda la lista tenga valores nulos
# y por lo tanto tenga que hacer iteraciones de toda la lista, la cantidad de iteración que habrán será la longitud de la lista, 
# oséase "n".
def primerValor(list):
    for x in list:
        if(x != None):
            print(x)
            return x
listA = [1]
tiempos=[]
valores=[]
for num in range(1,31):
    start=time.time()
    valor=primerValor(listA)
    end=time.time()
    final=end-start
    valores.append(valor)
    tiempos.append(final)
    listA.append(num)
print(tiempos)
indices = range(len(tiempos))
plt.bar(indices, tiempos)
```

```python
#la complejidad es o(n)^2 porque recorrere x*x veces mi lista para imprimir todas las combinaciones
def ej3(lst):
    for x in lst:
        for y in lst:
            print (""+str(x)+","+str(y))

listA = [1]
tiempos=[]
for num in range(1,31):
    start=time.time()
    ej3(listA)
    end=time.time()
    final=end-start
    tiempos.append(final)
    listA.append(num)
print(tiempos)

indices = range(len(tiempos))
plt.bar(indices, tiempos)
```
